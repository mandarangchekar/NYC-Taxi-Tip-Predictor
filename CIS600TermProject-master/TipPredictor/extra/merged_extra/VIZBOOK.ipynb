```python
import pandas as pd
import numpy as np
df = pd.read_csv("../../data/trip_merged_onemonth.csv")
```

```python
df[' tip_amount'].head()
```

```python
import matplotlib
matplotlib.use('Agg')
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(df[' passenger_count'][:100000], df[' tip_amount'][:100000], alpha=.5)
plt.ylabel("Tip Amount $")
plt.xlabel("Passenger Count")
plt.grid(b=True, which='major', axis='y')  
plt.title("Tip by passenger Count")
```

```python
plt.scatter(df[' trip_distance'][:100000], df[' tip_amount'][:100000], alpha=.5)
plt.ylabel("Tip Amount $")
plt.xlabel("Trip Distance miles")
plt.grid(b=True, which='major', axis='y')  
plt.title("Tip by Trip Distance") 
```

```python
plt.scatter(df[' trip_time_in_secs'][:100000], df[' tip_amount'][:100000], alpha=.5)
plt.ylabel("Tip Amount $")
plt.xlabel("Trip Time secs")
plt.grid(b=True, which='major', axis='y')  
plt.title("Tip by Trip time")
```

```python
# create a mapping from categorical values to unique integers:
integer_map = dict([(val, i) for i, val in enumerate(set(df[' payment_type']))]) 

# plot using this integer map, applied over `categorical`
ax = df[' payment_type'].apply(lambda x: integer_map[x])


plt.scatter(ax[:100000], df[' tip_amount'][:100000], alpha=.5)
plt.ylabel("Tip Amount $")
plt.xlabel("Payment Type ")
plt.grid(b=True, which='major', axis='y')  
plt.title("Tip by Payment Type")
```
